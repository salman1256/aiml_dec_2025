**Day-02 Exercise-02**

---
You are creating a simple library management system that stores book name in a list.
The program let you do the following things
1. Add Book
2. Display All books


---
Hint: Use list of books and while loop to do the above tasks




```python
books=['Basic of Python','AI from Start','Power BI for advanced Learners','You can win']
```

```python
while True:
  print('1. Add Book')
  print('2. Display All books')
  print('3. Exit')
  choice=input('Enter your choice(1-3): ')
  if choice=='1':
    book=input('Enter the book name: ')
    books.append(book)
    print('Book added successfully')
  elif choice=='2':
    print('All books in library are: ')
    for book in books:
      print(book)
  elif choice=='3':
    print('Exiting from Program')
    break
  else:
    print('Invalid Choice')
```
