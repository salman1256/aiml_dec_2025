```python
second_set={'Raj','Ravi','Sam','Amit'}
```

```python
print(second_set)
```

```python
for i in second_set:
    print(i,end=' ')
```

```python
# add item in set
second_set.add('Raju')
```

```python
print('Number of items after adding one items',len(second_set))
for item in second_set:
    print(item,end=' ')
```

```python
second_set.add('Nilesh')
```

```python
print('Number of items after adding one items',len(second_set))
for item in second_set:
    print(item,end=' ')
```

```python
# remove item from set
second_set.remove('Nilesh')
```

```python
second_set.remove('Sam')
```

```python
print('Number of items after removing 2 items',len(second_set))
for item in second_set:
    print(item,end='\t')
```

```python
second_set.remove('Sam')
```

```python
second_set.discard('Raju')
```

```python
print('Number of items after discard',len(second_set))
for item in second_set:
    print(item,end='\t')
```

```python
second_set.discard('Raju') #discord removes the item from set but no error if item not found inside set
```

```python
second_set.clear()
```

```python
print('Number of items after clear',len(second_set))
for item in second_set:
    print(item,end='\t')
```
