```python
product=(
    'Wireless Mike',
    'MVVI',
     400,
    4.6
)
```

```python
# A list  of employees using tuple (id,name,salary,designation)
employees=(
    (1,'John',50000,'Manager'),
    (2,'Jane',60000,'Developer'),
    (3,'Bob',55000,'Tester'),
    (4,'Alice',70000,'Designer'),
    (5,'Sam',50000,'Developer')

)
```

```python
for emp in employees:
  print(f'ID: \t {emp[0]}')
  print(f'Name: \t {emp[1]}')
  print(f'Salary: \t {emp[2]}')
  print(f'Designation: \t {emp[3]}')
  print('\n')


```
