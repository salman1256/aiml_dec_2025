* **break**: to break down a loop to end the loop
* **continue:**: skip that particular iteration and loop will continue after that

```python
for i in range(1,100):
  if(i%5==0):
    break
  print(i,end='\t')
```

```python
for i in range(1,100):
  if(i%5==0):
   continue
  print(i,end='\t')
```

```python
count=10
while count>0:
  print(count,end='\t')
  count=count-1  #count=count-1
```

```python
choice='y'
while(choice=='y'):
  given_number=int(input('Enter a number to find out table of number: \t'))
  print(f'Table of {given_number} as follows')
  for i in range(1,11):
     print(f'{given_number}*{i} = {given_number*i}')
  choice=input('Do you wanna continue..if yes press y to exit press any other key').lower()
```

```python
saved_username='admin'
saved_password='admin@123'
while True:
    username=input('Enter username: ')
    password=input('Enter password: ')
    if username==saved_username and password==saved_password:
        print('Login Successful')
        break

    else:
        print('Login Failed')
        print('Incorrect username or password Please try again')
        continue
```

```python
saved_username='admin'
saved_password='admin@123'
login_attempts=3
while login_attempts>0:
    username=input('Enter username: ')
    password=input('Enter password: ')
    if username==saved_username and password==saved_password:
        print('Login Successful')
        break

    else:
        print('Login Failed')
        login_attempts=login_attempts-1
        print(f'You have {login_attempts} attempts left')
        if login_attempts==0:
            print('You have exceeded the maximum number of attempts')
            print('Your account has been locked')
            break
```
