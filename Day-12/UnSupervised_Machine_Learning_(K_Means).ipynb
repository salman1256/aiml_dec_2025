* **Unsupervised Machine Learning** : NO target/output, The algorithm finds pattern by itself.
* We used unsupervised machine learning to finds hidden patterns,group similar data, and structure.

**K-means Clustering**
* Cluster: We will partition our dataset into K distinct group (Cluster)
* Centroid:Nearest mean cluster

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
dataset={
    "Entry_time":[9,10,11,14,15,16,17,18,19,20], #start browsing the website
    'Duration':[30,45,40,90,80,85,100,105,110,125]
}
```

```python
df=pd.DataFrame(dataset)
df
```

```python
df.shape
```

```python
model_kmeans=KMeans(n_clusters=3,random_state=42)
```

```python
df['cluster']=model_kmeans.fit_predict(df[['Entry_time','Duration']])
```

```python
df['cluster']
```

```python
centroid=model_kmeans.cluster_centers_
```

```python
plt.figure(figsize=(6,5))
plt.scatter(df['Entry_time'],df['Duration'],c=df['cluster'])
plt.scatter(centroid[:,0],centroid[:,1],s=100,color='red',marker='*')
plt.xlabel('Entry_time')
plt.ylabel('Duration')
plt.grid(True,linestyle='--',color='lightblue')
plt.show( )
```
