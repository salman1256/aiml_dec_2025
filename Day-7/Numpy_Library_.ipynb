```python
import numpy as np
```

```python
#Numpy array is called ndarray
array1=np.array([15,20,25,30,14,55,33])
print(array1)
```

```python
print('Type of array: \t',type(array1))
```

```python
print('Dimension of Array: \t',array1.ndim)
```

```python
print('Shape of Array: \t',array1.shape)
```

```python
print('Size of Array',array1.size)
```

```python
print('DataType of Array',array1.dtype)
```

```python
for (i,item) in enumerate(array1):
  print(i,item)
```

```python
print(f'Total of 3rd:{array1[2]} and 4th: {array1[3]} items is: \t ',(array1[2]+array1[3]))
```

**Numpy 2D Array**

```python
array2=np.array([[10,20,30,40],[240,350,450,550]])
```

```python
print(array2)
```

```python
print('array2 Dimension: \t',array2.ndim)
print('array2 Shape: \t',array2.shape)
print('array2 Size: \t',array2.size)
print('array2 Data Type: \t',array2.dtype)
print('array2 Item Size: \t',array2.itemsize)
```

```python
for i,v in np.ndenumerate(array2):
  print(i,v)
```

####NumPy can create Zero arrays in all dimension using np.zeros() with 0 values filled
* 1D Vector
* 2D Matrix
*  3D Array
* n dimensional array

```python
vector_zeros=np.zeros(10,dtype=float)
print(vector_zeros)
```

```python
print('Dimension: \t',vector_zeros.ndim)
print('Shape: \t',vector_zeros.shape)
print('Size: \t',vector_zeros.size)
print('Data Type: \t',vector_zeros.dtype)
```

```python
#numpy.zeros(shape,dtype=...)
matrix_z=np.zeros((3,4),dtype=int)
print(matrix_z)
```

```python
print('Dimension: \t',matrix_z.ndim)
print('Shape: \t',matrix_z.shape)
print('Size: \t',matrix_z.size)
print('Data Type: \t',matrix_z.dtype)
```

```python
# 3D Array
arr3=np.zeros((2,3,5),dtype=int)  # block,rows,columns
print(arr3)
```

```python
print('Dimension: \t',arr3.ndim)
print('Shape: \t',arr3.shape)
print('Size: \t',arr3.size)
print('Data Type: \t',arr3.dtype)
```

```python
arr4=np.zeros((2,3,4,5),dtype=int)
print(arr4)
```

```python
print('Dimension: \t',arr4.ndim)
print('Shape: \t',arr4.shape)
print('Size: \t',arr4.size)
print('Data Type: \t',arr4.dtype)
```

####NumPy can creat arrays in all dimension using np.ones() with values filled by 1

```python
# 1 D
arr_ones=np.ones(5,dtype=int)
print(arr_ones)
```

```python
print('Dimension: \t',arr_ones.ndim)
print('Shape: \t',arr_ones.shape)
print('Size: \t',arr_ones.size)
print('Data Type: \t',arr_ones.dtype)
```

```python
arr_2=np.ones((2,3),dtype=int)
print(arr_2)
```

```python
print('Dimension: \t',arr_2.ndim)
print('Shape: \t',arr_2.shape)
print('Size: \t',arr_2.size)
print('Data Type: \t',arr_2.dtype)
```

```python
arr3=np.ones((2,3,5),dtype=int)  # block,rows,columns
print(arr3)
```

```python
print('Dimension: \t',arr3.ndim)
print('Shape: \t',arr3.shape)
print('Size: \t',arr3.size)
```

```python
print('Dimension: \t',arr4.ndim)
print('Shape: \t',arr4.shape)
print('Size: \t',arr4.size)
print('Data Type: \t',arr4.dtype)
```

```python
array_constant=np.full((2,3,4,5),15)
print(array_constant)
```
