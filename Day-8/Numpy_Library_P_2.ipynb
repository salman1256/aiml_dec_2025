```python
#Sorting of NumPy array
my_sort_array=np.sort(myarray)
print(my_sort_array)
```

```python
# Array of All items greater than 100
array_greater_than_100=my_sort_array[my_sort_array>100]
print(array_greater_than_100)
#
```

```python
# array of items excluding 10 and 100
array_excluding_10_100=my_sort_array[(my_sort_array!=10) &(my_sort_array!=100)]
print(array_excluding_10_100)
```

```python
# array greater than 10 and less than 100
array_greater_than_10_less_than_100=my_sort_array[(my_sort_array>10) & (my_sort_array<100)]
print(array_greater_than_10_less_than_100)
#
```

```python
temp_kl=np.array([35,35,28,35,27,29,34,33,25,32,28,26,25,25,24,24,25,30,24,26,32,25,23,32,25,24,28,24,25,30
])
print(temp_kl)
```

```python
print('Size: \t',temp_kl.size)
```

```python
#Average temp. of kuala lumpur in last 30 days
#shortcut to open charmap Window + .
average_temp=np.mean(temp_kl)
print(f'Average temperature in KL: {round(average_temp,2)} °C')
#
```

```python
#Maximum temp
max_temp=np.max(temp_kl)
print(f'Maximum temperature in KL: {max_temp} °C')
```

```python
#Minimum Temp
min_temp=np.min(temp_kl)
print(f'Minimum temperature in KL: {min_temp} °C')
```

```python
hot_days=temp_kl[temp_kl>30]
print(f'Hot days: {hot_days}')
```

```python
print(f'Hot Days {hot_days.size} are following')
for day in hot_days:
  print(f'{day}°C', end='\t')
```

```python
cool_days=temp_kl[temp_kl<25]
print(f'Number of Cool Days {cool_days.size},Days are as follows')
for day in cool_days:
  print(f'{day}°C', end='\t')
```

```python
students_marks=np.array(
    [
        [90,80,80,80,80,90],
        [90,71,86,90,75,90],
        [98,88,86,90,89,90],
        [88,82,86,91,85,76],
        [68,81,82,90,71,75],

    ]
)

# * Herea Each Row=One Student
# Each Column =marks in Subject
# We have 5 Students record in 6 subject
```

```python
print(students_marks)
```

```python
print('Shape of student_marks',students_marks.shape)
```

```python
student_total_marks=np.sum(students_marks,axis=1)
print('Total marks of each student:')
print(student_total_marks)
#
```

```python
# Find out highest marks in each subject
highest_marks_in_subject=np.max(students_marks,axis=0)
print('Highest marks in each subjcets:')
print(highest_marks_in_subject)
```

```python
lowest_marks_in_subject=np.min(students_marks,axis=0)
print('Min marks in each subjcets:')
print(lowest_marks_in_subject)
```
