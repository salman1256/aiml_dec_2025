### Second Example of POS

```python
dataset={
    'text':['Please submit the report by Monday',
            'Send me the file ',
            'Kindly review the attached document',
            'Please process the resume as  soon as possible',
            'Bro share it',
            'Please confirm your result from officail website as well',
            'Why have not send the file yet'

    ],
    'label':[1,0,1,1,0,1,0] # 1 formal Conversation, 0 =Informal Conversation
}
```

```python
df=pd.DataFrame(dataset)
```

```python
df
```

```python
def pos_tagging(text):
    words=nltk.word_tokenize(text)
    pos_tags=nltk.pos_tag(words)
    counts=nltk.FreqDist(tag for word,tag in pos_tags)
    return counts
```

```python
df['features']=df['text'].apply(pos_tagging)
```

```python
df['features']
```

### NER: Named Entity Recognition (NER)
* Identifies names,dates,orgnizations,locations and others in text.


```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('maxent_ne_chunker_tab')
nltk.download('words')
```

```python
sentence='Sam John works at Apple in Calofornia and joined on 11 March 2013'
```

```python
# tokenization
tokens=nltk.word_tokenize(sentence)
tokens
```

```python
#pos tagging
pos_tags=nltk.pos_tag(tokens)
pos_tags
#
```

```python
#NER
ner=nltk.ne_chunk(pos_tags)
print(ner)
```

```python
list_ner=[]
for chunk in ner:
    if hasattr(chunk,'label'):
        entity=' '.join(c[0] for c in chunk)
        label=chunk.label()
        list_ner.append((entity,label))
print(list_ner)

```

* It helps machines to understand human languages efficienty.

```python
import spacy
```

```python
resume_text='Sam is a Senior Software Engineer with 10 years of experience. He worked at Microsoft and IBM in London.He has strong skills in Python Machine Learning , NLP, Deep Learning and Azure'
```

```python
nlp=spacy.load('en_core_web_sm')
```

```python
documents=nlp(resume_text)
documents
```

```python
#Named Entity Recognition NER
for entity in documents.ents:
    print(f'Entity: {entity.text},Type: {entity.label_}')
```

```python
#POS Tagging
for token in documents:
    print(f'Token: {token.text},POS: {token.pos_}')
```
