```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.model_selection import train_test_split
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
df=pd.read_csv('/content/sample_data/faq_100_records.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
#TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_faq=tfidf_vectorizer.fit_transform(df['Question'])
```

```python
def chatbot_response(user_question):
  question_vector=tfidf_vectorizer.transform([user_question])
  similarity_scores=cosine_similarity(question_vector,tfidf_faq)
  most_similar_index=np.argmax(similarity_scores)
  return df.iloc[most_similar_index]['Answer']
```

```python
your_question=input("Ask a question:")
response=chatbot_response(your_question)
print("Chatbot:",response)
```
