TensorFlow is an open-source end to end machine learning library for preprocessing data, modeling data and serving models.


```python
import tensorflow as tf
```

```python
print(tf.__version__)
```

```python
#Create Tensor with constant
scaler=tf.constant(7)
scaler
```

```python
print('Dimension: \t',scaler.ndim)
```

```python
print('Shape: \t',scaler.shape)
```

```python
print('Datatype: \t',scaler.dtype)
```

```python
vector=tf.constant([10,10])
vector
```

```python
print('Dimension: \t',vector.ndim)
print('Shape: \t',vector.shape)
print('Datatype: \t',vector.dtype)

```

```python
matrix=tf.constant([[10,7],[7,10]])
matrix
```

```python
print('Dimension: \t',matrix.ndim)
print('Shape: \t',matrix.shape)
print('Datatype: \t',matrix.dtype)
```

```python
import numpy
```

```python
size=tf.size(matrix)
print(size.numpy())
```

```python
print('Scaler Size: ',tf.size(scaler).numpy())
print('Vector Size: ',tf.size(vector).numpy())
print('Matrix Size: ',tf.size(matrix).numpy())
```
