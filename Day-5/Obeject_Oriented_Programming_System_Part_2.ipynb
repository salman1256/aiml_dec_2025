```python
class Product:
    def __init__(self, id,name, price):
        self.id = id
        self.name = name
        self.price = price
    def __str__(self):
      return f"Product Id:{self.id} \t Product Name: {self.name} \t Product Price: {self.price}"
```

```python
p1=Product(101,'Laptop',3000)
print(p1)
```

```python
p2=Product(102,'Mobile',1000)
p3=Product(103,'TV',2000)
p4=Product(104,'AC',4000)
p5=Product(105,'Fridge',5000)
```

```python
products=[p1,p2,p3,p4,p5]
for p in products:
  print(p)
```

We have  class Product  and its child class HerbalProduct.
* Product has these attributes id,name,price
* Herbal product has extra attributes herbs_used and expiry_date
* Use method overridding to regiter product and herbal product
* Use method overrriding to register,display product and herbal product details
* Create one one object of both classes & call both methods register & display


```python
#Product has these attributes id,name,price
class Product:
  def register(self):
    self.id=int(input("Enter id:"))
    self.name=input("Enter name:")
    self.price=float(input("Enter price:"))
  def display(self):
    print("Id:",self.id)
    print("Name:",self.name)
    print("Price:",self.price)
```

```python
class HerbalProduct(Product):
  def register(self):
    super().register()
    self.herbs_used=input("Enter herbs used:")
    self.expiry_date=input("Enter expiry date:")
  def display(self):
    super().display()
    print("Herbs used:",self.herbs_used)
    print("Expiry date:",self.expiry_date)
```

```python
p1=Product()
p1.register()
p1.display()
```

```python
hp1=HerbalProduct()
hp1.register()
hp1.display()
```
