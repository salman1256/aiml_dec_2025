```python
class Employee:
    def __init__(self,name,salary):
        self.name=name
        self.salary=salary

    def display(self):
      print('Employee Name: \t',self.name)
      print('Employee Salary \t',self.salary)
```

```python
class Product:
    def __init__(self, name, price,mfDate):
        self.name = name
        self.price = price
        self.mfDate = mfDate
    def details(self):
        print('Product Name: \t',self.name)
        print('Product Price: \t',self.price)
        print('Product Manufacture Date: \t',self.mfDate)
```

```python
class BankAccount:
  def __init__(self, balance):
       self.balance = balance

  def get_balance(self):
    return self.balance

  def deposit(self,amount):
    self.balance+=amount

  def withdraw(self,amount):
    if(self.balance>=amount):
      self.balance-=amount
      print('Transaction success')
    else:
      print('Insufficient Amount in Account')
      print('Transaction Declined')
```

```python
balance=int(input('Enter your balance: '))
account=BankAccount(balance)
print('Account created successfully')
while True:
  print('1. Deposit')
  print('2. Withdraw')
  print('3. Check Balance')
  print('4. Exit')
  choice=int(input('Enter your choice: '))
  if(choice==1):
      amount=int(input('Enter the amount to deposit: '))
      account.deposit(amount)
      print('Amount deposited successfully')
  elif(choice==2):
    amount=int(input('Enter the amount to withdraw'))
    account.withdraw(amount)
  elif (choice==3):
    print('Your balance is: ',account.get_balance())
  elif (choice==4):
    break
  else:
    print('Invalid choice')




```

```python
class Employee:
  def __init__(self, id,name,qual):
    self.id=id
    self.name=name
    self.qual=qual

  def __str__(self):
     return f'ID: {self.id} \nName: {self.name} \nQualification:{self.qual}'
```

```python
emp1=Employee(101,'Sam','MCA')
print(emp1)
emp2=Employee(102,'Tom','BTECH')
print(emp2)
```

```python
class Developer(Employee):
  def __init__(self,id,name,qual,project,domain):
    super().__init__(id,name,qual)
    self.project=project
    self.domain=domain

  def __str__(self):
    return f'ID: {self.id} \nName: {self.name} \nQualification:{self.qual} \n Project: {self.project} \n Domain: {self.domain}'
```

```python
dev_one=Developer(113,'Sameer','BTech','e-carnival','Java')
dev2=Developer(114,'Raj','MCA','ai-model-creation','Python')
```

```python
print(dev_one)
print(dev2)
```

### Create a Mobile class to store mobile phone details (brand,model, price) and display them using method.
* Create class Mobile
* Use Constructor( __ init __ ) to accept (brand,model,price)
* create method show_info() to print mobile information
* Create at least two objects and call methods



```python
class Mobile:
  def __init__(self,brand,model,price):
    self.brand=brand
    self.model=model
    self.price=price
  def show_info(self):
    print('Brand: \t',self.brand)
    print('Model: \t',self.model)
    print('Price: \t',self.price)
```

```python
obj1=Mobile('Samsung','A52',15000)
obj2=Mobile('Apple','Iphone 14',120000)
print('First Mobile Information')
obj1.show_info()
print('Second Mobile Information')
obj2.show_info()
```
