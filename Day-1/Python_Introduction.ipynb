## **Day-01** Exercise-01
---

1.  Take following details from the user
    * BasicSalary
    * HRA
    * TA
   
2.  Calculate
    NetSalary =BasicSalary +(BasicSalary * HRA%)+(BasicSalary * TA%)
3.  Print all values clearly




```python
basic_salary=float(input('Enter the basic salary:\t'))
hra=float(input('Enter the HRA %:\t'))
ta=float(input('Enter the TA %:\t'))
hra=(hra/100)*basic_salary
ta=(ta/100)*basic_salary
net_salary=basic_salary+hra+ta
print('Basic Salary: \t',basic_salary)
print('HRA: \t',hra)
print('TA: \t',ta)
print('Net Salary \t',net_salary)
```
