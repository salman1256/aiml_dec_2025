**Take following details from user**
* username, userpassword,is_blocked
* The username must be 'admin',
* Password must be 'admin@123'
* User must not be blocked

## if all three things correct 'Login Success'
## else 'Login Failed'


```python
user_name=input('Enter username :\t')
user_password=input('Enter password :\t')
is_blocked=bool(int(input('Enter is_blocked : 1 for yes 0 for No ')))
print('Your username is: ',user_name)
print('Your password is: ',user_password)
print('Your is_blocked is: ',is_blocked)
if(user_name=='admin' and user_password=='admin@123' and is_blocked==False):
  print('Login Success')
else:
  print('Login Failed')
```

```python
marks_per=float(input('Enter marks percentage :\t'))
if(marks_per>=90):
  print('Excellent')
  grade='A'
elif(marks_per>=70):
  print('Well Done')
  grade='B'
elif (marks_per>=50):
  print('Good')
  grade='C'
elif(marks_per>=40):
  print('Satisfactory')
  grade='D'
else:
  grade='F'
  print('Fail')
  grade='F'
print('Your marks percentage is: ',marks_per, '%')
print('Your grade is: ',grade)
```
