#Recursive functions:
* if function is calling to itself ,then the function is known as recursive function

```python
def factorial(num):
  if num==1:
    return 1
  else:
    return num*factorial(num-1)
```

```python
number=int(input('Enter a number to find out factorial: '))
print(f'The factorial of {number} is {factorial(number)}')
```

```python
def reversed_string(string):
  if len(string) <=1:
    return string
  else:
    return  string[-1]+reversed_string(string[:-1])
```

```python
reversed_string('welcome')
```

```python
reversed_string('hi')
```

```python
reversed_string('sam')
```

```python
mytext=input('Text to reverse: \t')
print(f'Original Text:\t {mytext} \nReversed Text:\t {reversed_string(mytext)}')
```

##### Write a recursive function that takes a number and returns sum of its digit

---
* e.g. numpber: 123 Sum of its digit {1+2+3} =6
### **Hints**
* define function sum_of_digits(number)
* number%10 to get last digit (extract the last digit +call function for remaining digits)
* number//10 to remove last digit
* base when number becomes 0


```python
def sum_of_digits(number):
  if number==0:
    return 0
  else:
   return (number%10)+sum_of_digits(number//10)
```

```python
my_num=int(input('Enter a number to find out sum of its digits: '))
print(f'The sum of digits of {my_num} is {sum_of_digits(my_num)}')
```

```python
def sum_of_digits(n):
    n=abs(n)
    if n==0:
       return 0
    return(n%10)+sum_of_digits(n//10)
```
