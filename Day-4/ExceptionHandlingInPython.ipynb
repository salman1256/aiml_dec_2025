```python
def add_num():
  num1=int(input("Enter First Number: "))
  num2=int(input("Enter Second Number: "))
  print(f'Result after adding{num1} and {num2} is : \t ',num1+num2)
```

```python
add_num()
```

```python
#To Handle Runtime error we use exception Handling
# It will help you to handle the error gracefully and end the program properly
```

```python
try:
   num1=int(input("Enter First Number: "))
   num2=int(input("Enter Second Number: "))
   print(f'Result after adding{num1} and {num2} is : \t ',num1+num2)

except:
  print("Invalid Input")

finally:
  print("End of Program")
```

```python
# try,except, finally, raise, else
#1.try: Code that might cause error
# 2. except: Code that runs when error occurs
# 3. finally: runs always (in case of erro or no error)
# 4. raise: if you want manually thrown an exception
# 5. else:Runs when no error
```

```python
def get_marks():
     marks=int(input("Enter Marks: "))
     if marks<0 or marks>100:
       raise ValueError("Invalid Marks must be between 0 to 100")
     return marks

```

```python
try:
  marks=get_marks()
except ValueError as ve:
  print('Value Error',ve)
else:
  print('Marks Saved: \t',marks)


```
