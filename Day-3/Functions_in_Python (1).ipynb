### Write a function to calculate the cube of number entered by user.
* Take the number from user and display following details
* for example Number: 5 Cube : 125
* for example Number: 2 Cube : 8

```python
def cube(number):
  print(f'Number: {number} Cube: {number**3}')
```

```python
cube(5)
```

```python
cube(2)
```

```python
cube(9)
```

```python
#default parameter
def reg(name,age,gender='male'):
  print('Registeration Success')
  print('Name:',name)
  print('Age:',age)
  print('Gender:',gender)
```

```python
#we can pass multiple default parameters in parameter list of function
#but default parameter must be last in parameter list
def add_num(num1,num2=10,num3=5, num4=20):
  result=num1+num2+num3+num4
  print(f'Result after adding {num1},{num2},{num3},{num4} is = \t {result}')
```

```python
add_num(10)
```

```python
add_num(10,20)
```

```python
add_num(45,12,13,11)
```

```python
#variable length parameters or arguments
def add_num(*num):
  total=0
  print('Total Number passed by you are:',len(num))
  for i in num:
        total=total+i
  print(f'Result after adding {num} is = \t {total}')
```

```python
add_num(12,11)
```

```python
add_num(10,3,5,6,8)
```

```python
add_num(45,11,34)
```

```python
# 4. keyword arguments **kwargs
def display(**kwargs):
  print(kwargs)
```

```python
display(name='sam',age=25,gender='male')
```

```python
display(id=12)
```

```python
display(name='sam',country='india')
```

```python
def hotel_room_booking(customer_name,room_no,**others):
 print('Booking Details as follows')
 print('Customer Name:',customer_name)
 print('Room No:',room_no)
 if others:
   print('Other Details')
   for key,value in others.items():
     print(key,':',value)
```

```python
hotel_room_booking('sam',123)
```

```python
hotel_room_booking('arsh',12,nights=5,breakfast='yes',lunch='yes')
```
