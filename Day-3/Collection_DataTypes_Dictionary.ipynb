**A dictionary is a collection datatype that stores data in key-value pairs.**
* mutable
* key must be unique
* value might be duplicate
* {'key':'value'}


```python
player={'name':'Lionel Messi','age':35,'country':'Argentina','ptype':'footballer'}
print(player)
#
```

```python
print('All Keys')
for k in player.keys():
  print(k)
```

```python
print('All Values')
for v in player.values():
  print(v)
```

```python
print('All Items')
for i in player.items():
  print(i)
```

```python
print('All Key-Value Pairs')
for k,v in player.items():
  print(k,v)
```

```python
print('Key \t Value ')
for k in player.keys():
  print(k,'\t',player[k])
```

```python
# Add new field to dictionary['key']=value
player['club']='PSG'
print(player)
```

```python
player['weight_in_kg']=67
```

```python
print('All Key-Value Pairs after changing')
for k,v in player.items():
  print(k,v)
```

```python
# update field in dictionary dictionary['key']=newvalue
player['club']='Real Madrid'
print(player)
```

```python
player['weight_in_kg']=68
print(player)
```

```python
player.pop('club') #delete key value pair from dictionary
```

```python
# key exist or not
search_key=input('Enter Key to Search: ')
if search_key in player:
  print('Key Exist')
  print(f'{search_key} is: ',player[search_key])
else:
  print('Key Not Exist')
```

```python
student={}
while True:
  print('\n ---Student Record System ---')
  print('1. Add Student')
  print('2. Display Detail')
  print('3. Update Detail')
  print('4. Delete Record')
  print('5. Exit')
  choice=int(input('Enter Your Choice(1-3): '))
  if choice==1:
    print('Add Student')
    student['name']=input('Enter Name: ')
    student['city']=input('Enter City: ')
    print('Student Added Successfully')
  elif choice==2:
   if student:
     print('Student Detail')
     print('Name: ',student['name'])
     print('City: ',student['city'])
   else:
     print('No Student Detail Found')
  elif choice==3:
    if student:
      field=input('Enter Field to Update: ')
      if field in student:
        student[field]=input(f'Enter New Value: for {field} ')
        print('Updated Successfully')
      else:
        print('Invalid Field')
    else:
      print('No Student Detail Found')

  elif choice==4:
    if student:
      student.clear()
      print('Record Deleted Successfully')
    else:
      print('No Record found')

  elif choice==5:
    print('The End')
    break
  else:
    print('\n Invalie Choice choose 1 to 3 only')


```

```python
product={'id':119,'name':'IPhone-16','price':5500.50} #A.
```

```python
product['company']='apple' #2.
```

```python
print('All Keys')      #4. Print all keys of the product dictionary
for k in product.keys():
  print(k)
```

```python
print('All Items')              # 5. Print all items of the product dictionary
for i in product.items():
  print(i)
```
