```python
#union,intersection, difference
set_one={10,20,30,40,50}
set_two={5,10,15,20,25,30}
```

```python
#union
set_union=set_one.union(set_two)
print('Union of both sets')
print(set_union)
```

```python
#intersection
set_intersection=set_one.intersection(set_two)
print('Intersection of both sets')
print(set_intersection)
```

```python
# difference
set_difference_1=set_one.difference(set_two)
print('Difference of both sets')
print(set_difference_1)
```

```python
set_difference_2=set_two.difference(set_one)
print('Difference of both sets')
print(set_difference_2)
#
```

```python
course_one={'Python','Power BI','Excel','SQL','Time Management','Soft Skill'}
course_two={'Python','Excel','Soft Skill','TesnorFlow','Azure','GitHub','Tableau'}
print(course_one)
print(course_two)
print('Number of subjects in course_one',len(course_one))
print('Number of subjects in course_two',len(course_two))
#
```

```python
common_subjects=course_one.intersection(course_two)
print('Common subjects in both courses')
for subject in common_subjects:
    print(subject)
```

```python
# if i take both the courses ,what all subjects ,i got , number of subjects
all_subjects=course_one.union(course_two)
print('Number of subjects covered: ',len(all_subjects))
print('All subjects in both courses')

for subject in all_subjects:
    print(subject)
```

```python
#The subjects Covered by course_one only ,(the subjects are in course_one but not in course_two) number of subjects
```

```python
course_one_only=course_one.difference(course_two)
print('Number of subjects covered by course_one only: ',len(course_one_only))
print('Subjects covered by course_one only')

for subject in course_one_only:
    print(subject)
```

```python
course_two_only=course_two.difference(course_one)
print('Number of subjects covered by course_two only: ',len(course_two_only))
print('Subjects covered by course_two only')

for subject in course_two_only:
    print(subject)
```
